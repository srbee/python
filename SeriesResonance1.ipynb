#Series Resonance

```
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt
pi=math.pi
# =========================================================
R=1
fr=2000 # Resonant Freq 
L=3e-3
C = (1)/(4*(pi**2)*L*(fr**2))
#==========================================================
print('Resonant Freq=',fr,'Hz','L=%6.2e'%L,'C=%6.2e'%C)
freq_range=(fr/5,fr*5,fr/200) # Frequency sweep
f=np.arange(*freq_range) # Note the *
w=2*pi*f # radian freq sweep
XL=w*L               # Inductive reactance values
XC=-1/(w*C)          # Capacitive reactance values
Z=R+XL+XC            # Complex impedance values
Z_mag=np.abs(Z)    # Magnitude of complex impedance values
plt.plot(f,Z_mag,'r',label='Z')
plt.plot(f,XL,'g',label='XL')
plt.plot(f,XC,'b',label='XC')
plt.title('Series Resonance')
plt.xlabel('Frequency in Hz')
plt.ylabel('XL,XC,Zmagtd')
plt.legend()
plt.grid()
```
