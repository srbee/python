## Automatic Generation of Characteristic Equation Given The Roots

```
import numpy as np
#========= Enter roots below ==============================
roots=np.array([1,2,3,4])    # coeffs=[1,-10,35 ,-50,24]
#roots=np.array([-1,2,-3,7]) # coeffs=[1,-5 ,-19,29 ,42]
#===============(o)(o)(o)==================================
n_roots=roots.size
m=np.empty(n_roots+1,dtype=np.object)
for i in range(m.shape[0]):
    m[i]=[]
    m[i].append(0)
print(m)
#============ Seed data =======================
m[0]=[0]
m[1]=[1,2,4]
m[2]=[3,5,6]
m[3]=[7]
#============== End of seed data ==============
n=3
print('n=',n)
for i in range(1,n+1,1):
    print('m[',i,']=',m[i])

n=n+1
while(n < n_roots+1):
 for i in range(n-1,0,-1):
  m[i].extend(x+pow(2,n-1) for x in m[i-1]) 

 m[n][0]=m[n-1][0]+pow(2,n-1) 
 print('=========================') 
 print('n= ',n)      
 for i in range(1,n+1,1):
  print('new m[',i,']=',m[i])
 n=n+1
# I have to create as many binary matrices as length of m[1],m[2],m[3],m[4]
#========== 29 Oct Ver 22 start =============
bb=np.empty(len(m),dtype=np.object)
print('bb=',bb)
print('shape of bb',bb.shape[0])
for i in range(len(m)):
 bb[i]=np.empty(len(m[i]),dtype=np.object)
 print('shape of bb[',i,'] shape[0]',bb[i].shape[0])

j=0
for i in range(bb[i].shape[0]):
  bb[j][i]=[]
#============ 29 Oct Ver 22 end =============
 
k=-1
for x in m[1]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb[1][k]=res
print('m[1]=',m[1])
 
k=-1
for x in m[2]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb[2][k]=res
print('========================================================')
print('m[2]=',m[2])
 
k=-1
for x in m[3]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb[3][k]=res
print('========================================================')
print('m[3]=',m[3])
 
k=-1
for x in m[4]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb[4][k]=res
print('========================================================')
print('m[4]=',m[4])

a=np.empty(len(m),dtype=np.object)
print('length of m =======',len(m))
#==== Generation of 'a' matrix for accumulation of products ===
for i in range(len(m)-1):
 a[i]=np.ones(len(m[i+1]),dtype=complex)
#==========================================
 #========= function kOutOfn() starts ========================
def kOutOfn(nn,matsize,bin_matrix,a_matrix,k,root_array):
 for j in range(matsize):
  for i in range(nn):
   if(bin_matrix[j][i]==1):
    a_matrix[k-1][j]=a_matrix[k-1][j]*root_array[i]
 summation=np.sum(a_matrix[k-1])
 return summation
#======== (o)(o)(o) =========
#=== creation of big 'm' array m[]=[m1,m2,m3,m4,..,m_n_roots]
for i in range(len(m)-1):
    m[i]=np.array(m[i+1])
#======== (o)(o)(0) ==========
sum1=sum2=sum3=sum4=complex(0,0)
sum=np.array([sum1,sum2,sum3,sum4])
nn=n
coeff=np.zeros([5],dtype=complex)
coeff[0]=1
n=roots.size
nn=n
#======= Entire work gets done below ============
for i in range(n):
 sum[i]=kOutOfn(nn,(m[i]).size,bb[i+1],a,i+1,roots)
 coeff[i+1]=sum[i]
#================================================
print('========================================')
print('The roots were :',end=' ')
for i in range(n_roots):
    if(i==n_roots-1):
     print(roots[i],end='\n ')
    else:
     print(roots[i],end=' ,')   
print('========================================')
print('Coeffs are :',end='')
for i in range(coeff.size):
 coeff[i]=coeff[i]*pow(-1,i)
 if(i==coeff.size-1):
  print(coeff[i],end=' ')
 else:
  print(coeff[i],end=', ')
print('')
print('================= Char Eqn ===================================')
print('==============================================================')
print('s^4+',coeff[1],'s^3+',coeff[2],'s^2+',coeff[3],'s^1+',coeff[4])
print('==============================================================')
```
