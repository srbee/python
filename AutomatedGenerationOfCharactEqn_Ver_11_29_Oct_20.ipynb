## Automatic Generation of Characteristic Equation Given The Roots

```
import numpy as np
n_roots=4 #================ change here
#arr=np.array([1,2,3,4])    # coeffs=[1,-10,35 ,-50,24]
roots=np.array([-1,2,-3,7]) # coeffs=[1,-5 ,-19,29 ,42]
m=np.empty(n_roots+1,dtype=np.object)
for i in range(m.shape[0]):
    m[i]=[]
    m[i].append(0)
print(m)
#============ Seed data =======================
m[0]=[0]
m[1]=[1,2,4]
m[2]=[3,5,6]
m[3]=[7]
#============== End of seed data ==============
n=3
print('n=',n)
for i in range(1,n+1,1):
    print('m[',i,']=',m[i])

n=n+1
while(n < n_roots+1):
 for i in range(n-1,0,-1):
  m[i].extend(x+pow(2,n-1) for x in m[i-1]) 

 m[n][0]=m[n-1][0]+pow(2,n-1) 
 print('=========================') 
 print('n= ',n)      
 for i in range(1,n+1,1):
  print('new m[',i,']=',m[i])
 n=n+1
#========== Generating Binary Matrices from m1=[1,2,4,8]
#n=5
#m=np.empty(n,dtype=np.object)
#for i in range(m.shape[0]):
#    m[i]=[]
#    m[i].append(0)
#m[0]=[0]
#m[1]=[1,2,4,8]
#m[2]=[3,5,6,9,10,12]
#m[3]=[7,11,13,14]
#m[4]=[15]
# I have to create as many binary matrices as length of m[1],m[2],m[3],m[4]
bb1=np.empty(len(m[1]),dtype=np.object)
for i in range(bb1.shape[0]):
    bb1[i]=[]
    bb1[i].append(0)
k=-1
for x in m[1]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb1[k]=res
print('m[1]=',m[1])
for i in range(len(bb1)):
 print('bb1[',i,']=',bb1[i])
 #==============================================================
 bb2=np.empty(len(m[2]),dtype=np.object)
for i in range(bb2.shape[0]):
    bb2[i]=[]
    bb2[i].append(0)
k=-1
for x in m[2]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb2[k]=res
print('========================================================')
print('m[2]=',m[2])
for i in range(len(bb2)):
 print('bb2[',i,']=',bb2[i])
 #==============================================================
 bb3=np.empty(len(m[3]),dtype=np.object)
for i in range(bb3.shape[0]):
    bb3[i]=[]
    bb3[i].append(0)
k=-1
for x in m[3]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb3[k]=res
print('========================================================')
print('m[3]=',m[3])
for i in range(len(bb3)):
 print('bb3[',i,']=',bb3[i])
 #==============================================================
 bb4=np.empty(len(m[4]),dtype=np.object)
for i in range(bb4.shape[0]):
    bb4[i]=[]
    bb4[i].append(0)
k=-1
for x in m[4]:
    k=k+1
    first=x
    b1=bin(first)
    c=[]
    for digit in b1:
        c.append((digit))
    res=[int(ele) if ele.isdigit() else ele for ele in c]
    del res[:2]
    for i in range(4-len(res)):
        res.insert(0,0)
    bb4[k]=res
print('========================================================')
print('m[4]=',m[4])
for i in range(len(bb4)):
 print('bb4[',i,']=',bb4[i])
 #==== for putting together the characteristic eqn
 #==== The a matrices for accumulating sums of products taken n at a time
 a1=np.ones(len(m[1]),dtype=complex)
 a2=np.ones(len(m[2]),dtype=complex)
 a3=np.ones(len(m[3]),dtype=complex)
 a4=np.ones(len(m[4]),dtype=complex)
 a=np.array([a1,a2,a3,a4])
 #==========================================
 #========= function kOutOfn() starts ========================
def kOutOfn(nn,matsize,bin_matrix,a_matrix,k,root_array):
 for j in range(matsize):
  for i in range(nn):
   if(bin_matrix[j][i]==1):
    a_matrix[k-1][j]=a_matrix[k-1][j]*root_array[i]
 summation=np.sum(a_matrix[k-1])
 return summation
#============ function kOutOfn()  ends ========================
m1=np.array(m[1])
m2=np.array(m[2])
m3=np.array(m[3])
m4=np.array(m[4])
m=np.array([m1,m2,m3,m4])
binary=np.array([bb1,bb2,bb3,bb4])#=== change reqd
sum1=sum2=sum3=sum4=complex(0,0)
sum=np.array([sum1,sum2,sum3,sum4])
nn=n
coeff=np.zeros([5],dtype=complex)
coeff[0]=1

#arr=np.array([1,2,3,4])
#arr=np.array([-1,2,-3,7])
n=roots.size
nn=n
#======= Entire work gets done below ============
for i in range(n):
 sum[i]=kOutOfn(nn,(m[i]).size,binary[i],a,i+1,roots)
 coeff[i+1]=sum[i]
#================================================
print('Coeffs are =>',end='')
for i in range(coeff.size):
 coeff[i]=coeff[i]*pow(-1,i)
 if(i==coeff.size-1):
  print(coeff[i],end=' ')
 else:
  print(coeff[i],end=', ')
print('')
print('======== Char Eqn =========')
print('============================')
print('s^4+',coeff[1],'s^3+',coeff[2],'s^2+',coeff[3],'s^1+',coeff[4])
print('==============================================================')
```
