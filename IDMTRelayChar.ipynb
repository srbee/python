Relay Characteristics


```
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath
e=math.exp(1)
#pi=math.pi
#print(pi,e)
psm=np.linspace(2,20,50)
t_IDMT=(0.14)/((psm**0.02)-1) # This works
t_Inverse=(13.2)/((psm)-1)
t_ExtInv=(80)/((psm**2)-1)
#t=(0.14)/((math.pow(psm,0.02))-1) # This is not working
#=========================================================
x = np.linspace(2,20,10)


fig = plt.figure()
ax = fig.gca()
ax.set_xticks(x)
ax.set_yticks(np.arange(0,80 , 2))

#=========================================================

plt.plot(psm,t_IDMT,'r-.',label='IDMT')
plt.plot(psm,t_Inverse,'g-.',label='Inverse')
plt.plot(psm,t_ExtInv,'b-.',label='Ext Inverse')
plt.legend()
plt.xlabel('Plug Setting Multiplier')
plt.ylabel('Relay Operating Time in seconds  with TMS=1')
plt.title('Inverse Time Relay Characteristics for TMS=1')
plt.grid()
#print(t)
```
