```
# DC offset in R-L Circuit Driven by AC Supply switched at various points on the wave
import math
import numpy as np
import matplotlib.pyplot as plt
pi=math.pi
R=3
X=4
Z=R+X*1j
phi=math.atan2(X,R)
phi_deg=math.degrees(phi)
f=50
w=2*pi*f
Vm=100
cycles=5
T=1/f
dT=1e-3 # Sampling interval
t=[tim for tim in np.arange(0,cycles*T,dT)]
t=np.array(t,'float') # Rem t is an array
wt=w*t # Rem wt is an array
#print(time)
v=np.sin(wt) # Note: You have to use np.sin() and NOT math.sin() 
# since this is an array operation
plt.plot(v)
#print(R,X,Z)
print('phi=',phi,'phi_deg=',phi_deg)
```
